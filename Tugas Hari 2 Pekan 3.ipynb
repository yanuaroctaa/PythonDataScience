Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
plt.plot(x,cos, color='red',label='Cosx')
plt.plot(x,cos3, color='green', linestyle='dashed',label='3 Cosx')
plt.plot(x,sin, color='blue', linestyle='dashdot',label='Sin(x-90)')
plt.plot(x,sin2,marker='x', color='yellow',linestyle='none',label='2 Sinx')
plt.legend(loc='center')
ax.set_title('Grafik Cos Sin')
ax.set_xlabel('X.Axis Label')
ax.set_ylabel('Y.Axis Label')
plt.show()
```
